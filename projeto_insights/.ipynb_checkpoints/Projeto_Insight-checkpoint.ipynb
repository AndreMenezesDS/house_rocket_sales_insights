# 0.2 Funções de operação

```python
warnings.filterwarnings ('ignore')

def jupyter_settings():
    %matplotlib inline
    %pylab inline
    
    plt.style.use( 'bmh' )
    plt.rcParams['figure.figsize'] = [25, 12]
    plt.rcParams['font.size'] = 24
    
    display( HTML( '<style>.container { width:90% !important; }</style>') )
    pd.options.display.max_columns = None
    pd.options.display.max_rows = None
    pd.set_option( 'display.expand_frame_repr', False )
    
    sns.set()

def show_dspecs(data):
    dtypes = data.dtypes
    print('Number of rows: {}'.format(data.shape[0]))
    print('Number of columns:{}'.format(data.shape[1]))
    
    return dtypes

def descriptive_statistics( data ):
    num_attributes = data.select_dtypes( include=['int64', 'float64'] )

    # Métricas de tendência central
    media = pd.DataFrame( num_attributes.apply( np.mean ) )
    mediana = pd.DataFrame( num_attributes.apply( np.median ) )
    std = pd.DataFrame( num_attributes.apply( np.std ) )

    # Métricas de Dispersão
    std = pd.DataFrame( num_attributes.apply( np.std ) )
    max_ = pd.DataFrame( num_attributes.apply( np.max ) )
    min_ = pd.DataFrame( num_attributes.apply( np.min ) )

    df1 = pd.concat( [max_, min_, media, mediana, std ], axis=1 ).reset_index()
    df1.columns = ['attributes', 'mean', 'median', 'std', 'max', 'min']
    
    return df1

def add_columns(data):
    data['mes'] = data['date'].dt.month
    data['ano'] = data['date'].dt.year
    
    data['construcao'] = data['yr_built'].apply(lambda x: '< 1955' if (x < 1955) else
                                          '> 1955')
    
    data['season'] = data['mes'].apply(lambda x: 'verão' if (x >= 6) & (x <= 8) else
                                      'primavera' if (x >= 3) & (x <= 5) else
                                      'outono' if (x >= 9) & (x <= 11) else
                                      'inverno')
    
    data['banheiro_agrupamento'] =  data['bathrooms'].apply(lambda x: '0-3' if (x > 0 ) & (x < 3) else
                                                   '3-5' if (x > 3) & (x < 5) else
                                                   '5-8')
    
    data['area_media'] = data['sqft_living'].apply(lambda x: 'acima da média' if (x >= 13540.00) else 'abaixo da média')
    
    data['renovacao'] = data['yr_renovated'].apply(lambda x: 'sim' if x > 0 else
                                                     'nao'   )
    
    data['vista_agrupamento'] =  data['view'].apply(lambda x: '0-2' if (x > 0 ) & (x < 2) else
                                                   '3-4')
    data['avaliacao'] =  data['grade'].apply(lambda x: '0-6' if (x > 0 ) & (x <= 6) else
                                                   '7-13')
    
    data['porao'] = data['sqft_basement'].apply(lambda x: 'nao' if x == 0 else
                                               'sim')
    
    return data

def draw_map( data , filtra_zip):
    # map
    houses = data[(data['pay'] == filtra_zip)][['id', 'lat', 'long', 'condition', 'pay', 'price']] 
    fig = px.scatter_mapbox( houses, 
                             lat="lat", 
                             lon="long", 
                             color="condition", 
                             size="price",
                             color_continuous_scale=px.colors.cyclical.IceFire,
                             opacity = 0.5,
                             size_max=15, 
                             zoom=10)
    
    fig.update_layout(mapbox_style="open-street-map")
    fig.update_layout(height=600, margin={"r":0,"t":0,"l":0,"b":0})
    fig.show()

#Função auxiliar de formatação da visualização dos dados
pd.options.display.float_format = '{:.2f}'.format

def get_geofile(url):
    geofile = geopandas.read_file(url, timeout = False)
    
    return geofile


#Requisição do arquivo com o mapa
url = "https://opendata.arcgis.com/datasets/83fc2e72903343aabff6de8cb445b81c_2.geojson"
geofile = get_geofile(url)

def region_overview(data, geofile):
    df = data.sample(
        50
    )  # Aqui, é utilizado a amostragem apenas para rodar o código + rápido ;)

    density_map = folium.Map(
        location=[data["lat"].mean(), data["long"].mean()], default_zoom_start=15
    )

    marker_cluster = MarkerCluster().add_to(density_map)
    for name, row in df.iterrows():
        folium.Marker(
            [row["lat"], row["long"]],
            popup="Sold R${0} on {1}. Features: {2} sqft, {3} bedrooms, {4} bathrooms, year built: {5}".format(
                row["price"],
                row["date"],
                row["sqft_living"],
                row["bedrooms"],
                row["bathrooms"],
                row["yr_built"],
            ),
        ).add_to(marker_cluster)

    density_map

    # Region Price Map

    df = data[["price", "zipcode"]].groupby("zipcode").mean().reset_index()
    df.columns = ["ZIP", "PRICE"]
    df = df.sample(50)

    geofile = geofile[geofile["ZIP"].isin(df["ZIP"].tolist())]

    region_price_map = folium.Map(
        location=[data["lat"].mean(), data["long"].mean()], default_zoom_start=15
    )

    region_price_map.choropleth(
        data=df,
        geo_data=geofile,
        columns=["ZIP", "PRICE"],
        key_on="feature.properties.ZIP",
        fill_color="YlOrRd",
        fill_opacity=0.7,
        line_opacity=0.2,
        legend_name="AVG PRICE",
    )

    region_price_map

    return density_map, region_price_map
```

# 3. Questão de Negócio

## 3.1 Quais são os imóveis que a House Rocket deveria comprar e por qual preço?

```python
df = data.copy()
df = add_columns(df)
```

### 3.1.1. Quais Casas?

```python
median_column = df[['zipcode', 'price']].groupby('zipcode').median().reset_index()

df2 = pd.merge(median_column, df, on='zipcode', how='inner')
df2 = df2.rename( columns = {'price_y': 'price', 'price_x': 'price_median'})

for i, row in df2.iterrows():
    if (row['price_median'] >= row['price'] ) & (row['waterfront'] == 0) & (row['porao'] =='nao') & (row['avaliacao'] == '0-6') & (row['renovacao'] == 'nao') & ((row['mes'] <= 4) | (row['mes'] >= 10)):
        df2.loc[i, 'pay'] = 'sim'
    else:
        df2.loc[i, 'pay'] = 'nao'
        
```

```python
#Casas para comprar
compra = df2[df2['pay'] == 'sim']

compra['pay'].shape[0]
compra[['id', 'price', 'pay']]

```

```python
filtra_zip = widgets.Dropdown(options = df2['pay'].unique().tolist(),
                       description = 'Pay',
                       disabled = False)

widgets.interactive( draw_map, data = fixed(df2), filtra_zip  = filtra_zip)
```

### 3.2 Qual o melhor momento de venda para os imóveis comprados? E por qual preço?


```python
df3 = df2.copy()
df3 = df3[df3['pay'] == 'sim']

median_price = df3['price'].median()
median_price

for i, row in df3.iterrows():
    if (median_price > row['price']):
        df3.loc[i, 'sale'] =  row['price'] * 1.1
    else:
        df3.loc[i, 'sale'] = row['price'] * 1.3
        
```
